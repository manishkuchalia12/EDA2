Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how
they may impact your analysis.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'flight_price_dataset.csv' is the name of your dataset
flight_data = pd.read_csv('flight_price_dataset.csv')

# Create a boxplot to identify potential outliers
plt.figure(figsize=(12, 8))
sns.boxplot(x='Airline', y='Price', data=flight_data, palette='viridis')
plt.title('Flight Prices by Airline (with Outliers)')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

Identifying Outliers:
Outliers Above Upper Whisker: Points above the upper whisker of the boxplot may be considered outliers.
Outliers Below Lower Whisker: Points below the lower whisker of the boxplot may also be considered outliers.
Impact on Analysis:
Statistical Measures: Outliers can skew statistical measures like the mean and standard deviation.
Model Performance: Outliers can impact the performance of machine learning models, especially those sensitive to extreme values.
Assumptions: Outliers may violate assumptions of certain statistical analyses.
Dealing with Outliers:
Remove Outliers: You may choose to remove extreme values, but be cautious as this can influence results.
Transform Data: Apply transformations (e.g., log transformation) to reduce the impact of outliers.
Use Robust Statistics: Utilize robust statistical measures that are less sensitive to outliers.
Modeling Techniques: Some models are less affected by outliers (e.g., tree-based models).

Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may
impact your analysis.

```python
import pandas as pd

# Assuming 'google_playstore.csv' is the name of your dataset
playstore_data = pd.read_csv('google_playstore.csv')

# Check for missing values
missing_values = playstore_data.isnull().sum()

# Display the count of missing values for each column
print("Missing Values:\n", missing_values)

# Display the percentage of missing values for each column
percentage_missing = (missing_values / len(playstore_data)) * 100
print("\nPercentage of Missing Values:\n", percentage_missing)
```

Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
the relationship.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'google_playstore.csv' is the name of your dataset
playstore_data = pd.read_csv('google_playstore.csv')

# Drop rows with missing values in 'Size' or 'Rating' columns
playstore_data = playstore_data.dropna(subset=['Size', 'Rating'])

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(playstore_data['Size'], playstore_data['Rating'], alpha=0.5, color='blue')
plt.title('Relationship between App Size and Rating')
plt.xlabel('App Size (MB)')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'google_playstore.csv' is the name of your dataset
playstore_data = pd.read_csv('google_playstore.csv')

# Drop rows with missing values in 'Type' or 'Price' columns
playstore_data = playstore_data.dropna(subset=['Type', 'Price'])

# Convert 'Price' column to numeric, removing dollar signs
playstore_data['Price'] = playstore_data['Price'].apply(lambda x: float(x[1:]) if '$' in x else 0)

# Create a bar chart for average prices by app type
average_prices_by_type = playstore_data.groupby('Type')['Price'].mean()

plt.figure(figsize=(10, 6))
average_prices_by_type.plot(kind='bar', color=['skyblue', 'lightgreen'], edgecolor='black')
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.

```python
import pandas as pd

# Assuming 'google_playstore.csv' is the name of your dataset
playstore_data = pd.read_csv('google_playstore.csv')

# Drop rows with missing values in 'Installs' column
playstore_data = playstore_data.dropna(subset=['Installs'])

# Convert 'Installs' column to numeric, removing commas and plus signs
playstore_data['Installs'] = playstore_data['Installs'].apply(lambda x: int(x.replace(',', '').replace('+', '')))

# Create a frequency table for the top 10 most popular apps
top_apps = playstore_data.nlargest(10, 'Installs')[['App', 'Installs']]

print("Top 10 Most Popular Apps (by Installs):\n", top_apps)
```
